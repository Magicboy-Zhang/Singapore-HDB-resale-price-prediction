### Get new features related to famours primary school, using API from OneMap

### Data from https://elite.com.sg/primary-schools

```python
# data from https://elite.com.sg/primary-schools
# phase 2B - ranked by 2B%
# phase 2C - ranked by 2C%

# load txt files
file_B = open('primary_phase2B.txt')
file_C = open('primary_phase2C.txt')

phaseB = file_B.readlines()
phaseC = file_C.readlines()

# 181 primary schools in total, already sorted by %
print(phaseB)
print(len(phaseB))
print(phaseC)
print(len(phaseC))
```

```python
# get top 60 fron both phase B & C
top60_pb = phaseB[: 60]
top60_pc = phaseC[: 60]

print(top60_pb)
print(top60_pc)
```

```python
print(top60_pb[0].split('\t')[0])
```

```python
best_school = []

# pick the top 60 school according to 2B% & 2C%
for info in top60_pb:
    current_school = info.split('\t')[0]
    if current_school not in best_school:
        best_school.append(current_school)

for info in top60_pc:
    current_school = info[0].split('\t')[0]
    if current_school not in best_school:
        best_school.append(current_school)

print(best_school)
print(len(best_school))

# only store top 60
best_school = best_school[: 60]
```

```python
import requests as re
import pandas as pd

# hash table
top60_dict = {
    'School_name': [],
    'SearchVal': [],
    'lat': [],
    'lng': []
}

headers = {
    'purpose': 'student\'s project',
    'University': 'NTU'
}

for school in best_school:

    url = "https://www.onemap.gov.sg/api/common/elastic/search?searchVal=" + school + "&returnGeom=Y&getAddrDetails=Y&pageNum=1"

    # get a dictionary of all details
    response = re.request("GET", url, headers = headers).json()

    for result in response['results']:

        top60_dict['School_name'].append(school)
        top60_dict['SearchVal'].append(result['SEARCHVAL'])
        top60_dict['lat'].append(result['LATITUDE'])
        top60_dict['lng'].append(result['LONGITUDE'])

# for each row, get corresponded geolocation
top60_df = pd.DataFrame(top60_dict)

top60_df.to_csv('top60_primary.csv')

top60_df
```

```python
import math

'''
Calculate the E_dis
'''
def euclidean_dis(lat1, lng1, lat2, lng2):
    return math.sqrt((lat2 - lat1) ** 2 + (lng2 - lng1) ** 2)

primary_dict = {}
'''
move all useful data from 'top60_wo_df' into a dictionary - (lat, lng) : school_name
'''
def setup_primary_dict(row):
    global primary_dict
    primary_dict[(row['lat'], row['lng'])] = row['School_name']

    return row

# set up the dictionary for mrt geolocation
top60_wo_df.apply(setup_primary_dict, axis = 1)

'''
Function used in apply(), get minimum distance and construct new columns in 'hdb_df'
    - 'min_primary_lat'
    - 'min_primary_lng'
    - 'min_dis_primary_euclidean'
    - 'min_dis_primary'
'''
def get_miniPrimary_dis(row):
    global primary_dict
    min_dis = 999999
    min_dis_primary = None

    for coord in primary_dict.keys():
        current_dis = euclidean_dis(row['lat'], row['lng'], coord[0], coord[1])
        if current_dis < min_dis:
            min_dis = current_dis
            min_dis_primary = primary_dict[(coord[0], coord[1])]

            row['min_primary_lat'] = coord[0]
            row['min_primary_lng'] = coord[1]
    
    row['min_dis_primary_euclidean'] = min_dis
    row['min_dis_primary'] = min_dis_primary

    return row

hdb2023_df = pd.read_csv("2023_resale_price_with_geolocation_&_minMrt_w&t.csv")

hdb2023_df = hdb2023_df.apply(get_miniPrimary_dis, axis = 1)

hdb2023_df.to_csv('2023_resale_price_with_geolocation_&_minPrimary.csv', index = False)

hdb2023_df.head()
```
